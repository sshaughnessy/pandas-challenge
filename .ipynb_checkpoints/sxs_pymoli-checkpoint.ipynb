```python
#import libraries
import pandas as pd
```

```python
#resource file path
file = "HeroesOfPymoli/Resources/purchase_data.csv"
```

```python
#read csv file
purchase_df = pd.read_csv(file)
```

```python
#see what the data looks like
#purchase_df.head()
```

```python
#basic summary metrics of data
#purchase_df.describe()
```

```python
#variable holding total unique players
total_players = purchase_df["SN"].nunique()
#create summary data frame to display total unique players
total_players_df = pd.DataFrame({"total_players": [total_players]})
#display summary df
total_players_df
```

```python
#variable holding count of unique games 
unique_games = purchase_df["Item Name"].nunique()
#variable holding average price of all games
avg_price = purchase_df["Price"].mean()
#variable holding total number of purchases
num_purchases = purchase_df["Purchase ID"].count()
#variable holding sum of all sales
total_rev = purchase_df["Price"].sum()
#create summary df with above variables as columns
summary_df = pd.DataFrame({"unique_games": [unique_games], "average_price": [avg_price], "total_purchases": [num_purchases], "total_revenue": [total_rev]})
#display summary df
summary_df
```

```python
#group the df by gender
gender_group = purchase_df.groupby(["Gender"]) 
#from the grouped df, count unique players
gender_uniques = gender_group["SN"].nunique()
#from the grouped df, calc % of totalk
gender_percentage= gender_uniques/total_players
#create summary df
gender_df = pd.DataFrame({"players":gender_uniques,"percentage_of_total":gender_percentage})
#display summary df
gender_df
```

```python
#use the grouped by gender df object to count purchases
purchase_count = gender_group["Purchase ID"].nunique()
#average price of all games purchased in group
avg_price = gender_group["Price"].mean()
#sum of purchase price by group
total_spend = gender_group["Price"].sum()
#average spend per person in group
avg_per = total_spend/purchase_count
#create summary df
gender_demo_df = pd.DataFrame({"purchase_count":purchase_count,"average_price":avg_price,"total_purchase_value":total_spend,"avg_total_per_person":avg_per})
#display summary df
gender_demo_df
```

```python
#create bins and labels
bins = [0,9,14,19,24,29,34,39,200]
age_groups = ["<10","10-14", "15-19", "20-24","25-29","30-34","35-39","40+" ]
#add a column to df with age buckets
purchase_df["age_group"]=pd.cut(purchase_df["Age"], bins, labels = age_groups,right = True, include_lowest = True)
#groupby age buckets
age_groupby = purchase_df.groupby(["age_group"])
#perform calculations
age_counts = age_groupby["SN"].nunique()
age_percentage_total = age_counts/total_players
#new summary dataframe
age_demo_df = pd.DataFrame({"players":age_counts, "percent_of_total":age_percentage_total})
age_demo_df
```

```python
#use age groupby df object to calc # purchases
age_purchases = age_groupby["Purchase ID"].count()
#get average purchase price by age group
age_avg_price = age_groupby["Price"].mean()
#total purchase dollars by age group
age_total_spend = age_groupby["Price"].sum()
#average spend per person by age group
age_avg_spend = age_purchases/age_total_spend
#create summary df
age_analysis_df = pd.DataFrame({"purchases":age_purchases, "average_purchase_price":age_avg_price, "total_purchases":age_total_spend, "avg_spend_per_person":age_avg_spend})
#display summary df
age_analysis_df

```

```python
#create groupby object to group by gamer unique id
spend_groupby = purchase_df.groupby(["SN"])
#calc total purchase $ by gamer unique id
spend_total = spend_groupby["Price"].sum()
#count total purchases by gamer unique id
spend_count = spend_groupby["Purchase ID"].count()
#calc avg purchase price by gamer unique id
spend_avg = spend_total / spend_count
#create summary df
spend_summary_df = pd.DataFrame({"purchase_count":spend_count,"average_purchase_price":spend_avg,"total_purchase_value":spend_total})
#sort summary df by most to least purchases
spend_summary_df = spend_summary_df.sort_values(by=["total_purchase_value"], ascending=False)
#display summary
spend_summary_df.head()
```

```python
#create new df with selected columns
popular_df = purchase_df[["Item ID","Item Name","Price"]].copy()
#groupby item id and name
popular_group = popular_df.groupby(["Item ID", "Item Name"])
#calculate grouped purchase count
popular_count = popular_group["Item ID"].count()
#calculate grouped purchase value
popular_spend = popular_group["Price"].sum()
#calculate grouped purchase average
popular_avg_spend = popular_spend/popular_count
#create summary df
pop_summ_df = pd.DataFrame({"purchase_count":popular_count,"average_spend":popular_avg_spend,"total_spend":popular_spend})
#sort by purchase count
pop_summ_df = pop_summ_df.sort_values(by=["purchase_count"], ascending=False)
#display summary
pop_summ_df.head()
```
