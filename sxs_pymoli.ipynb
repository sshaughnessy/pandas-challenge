```python
import pandas as pd
```

```python
file = "HeroesOfPymoli/Resources/purchase_data.csv"
```

```python
purchase_df = pd.read_csv(file)
```

```python
purchase_df.head()
```

```python
purchase_df.describe()
```

```python
total_players = purchase_df["SN"].nunique()
total_players_df = pd.DataFrame({"total_players": [total_players]})
total_players_df
```

```python
unique_games = purchase_df["Item Name"].nunique()
avg_price = purchase_df["Price"].mean()
num_purchases = purchase_df["Purchase ID"].count()
total_rev = purchase_df["Price"].sum()
summary_df = pd.DataFrame({"unique_games": [unique_games], "average_price": [avg_price], "total_purchases": [num_purchases], "total_revenue": [total_rev]})
summary_df
```

```python
gender_group = purchase_df.groupby(["Gender"]) #group by object, not df
gender_uniques = gender_group["SN"].nunique()
gender_df = pd.DataFrame(gender_uniques)
gender_df["gender_percent"]= gender_df["SN"]/total_players
gender_df.rename(columns={"SN": "total_count"}) #why won't this rename?
gender_df
```

```python
gender_demo=purchase_df.groupby("Gender")
purchase_count = gender_demo["Purchase ID"].nunique()
avg_price = gender_demo["Price"].mean()
total_spend = gender_demo["Price"].sum()
gender_demo_df = pd.DataFrame(purchase_count)
gender_demo_df["avg_price"] = avg_price
gender_demo_df["total_purchase_value"] = total_spend
gender_demo_df["avg_total_per_person"] = total_spend/purchase_count
gender_demo_df.rename({"Purchase ID": "purchase_count"},axis=1) #why won't this rename? ##how do i get the groupby to a new dataframe without that first column
gender_demo_df
```
